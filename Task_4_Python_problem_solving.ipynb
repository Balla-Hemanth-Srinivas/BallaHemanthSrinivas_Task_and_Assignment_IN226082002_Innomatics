# Problem 1: Mobile Recharge Validation System

## Problem Statement:

Create a function to validate a mobile recharge request.
## Rules:

Recharge amount must be ≥ ₹50

Valid plans: 199, 299, 399, 599

User should be allowed to retry if invalid

## Real-Time Use:

Telecom recharge apps

Payment gateways


## Hint
Use function for validation

Use while loop for retry


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and amount in valid_plans:
        return True
    return False

while True:
    amt = int(input("Enter recharge amount: ₹"))
    if validate_recharge(amt):
        print(f"Recharge of ₹{amt} accepted ✔")
        break
    else:
        print(f"₹{amt} is an invalid amount. Try again!")
```

# Problem 2: Inventory Reorder Alert System

## Problem Statement:
Given a dictionary of products and stock quantities:

Print Reorder Alert if stock < 15.

Print Stock OK otherwise.


## Real-Time Use:
Warehouse systems.

Retail management.

## Hint:

Use function.

Use for loop on dictionary.

```python
def check_inventory(stock_dict):
    for product, qty in stock_dict.items():
        if qty < 15:
            print(f"{product}: Reorder Alert!")
        else:
            print(f"{product}: Stock OK")

products = {
    "Pens": 10,
    "Notebooks": 20,
    "Staplers": 5,
    "Markers": 18
}

check_inventory(products)
```

# Problem 3: Student Result Processing System

## Problem Statement:

Create a function that:

Accepts a list of student marks.

Calculates average marks.

Prints Pass if average ≥ 50 else Fail.

## Real-Time Use:

Online exam systems.

Evaluation portals.

## Hint

Use loop to calculate total.

Return result from function.

```python
def process_results(marks_list):
    total = 0
    for m in marks_list:
        total += m
    avg = total / len(marks_list)
    
    if avg >= 50:
        return "Pass"
    return "Fail"

students = [60, 45, 78, 82, 50]
print("Result:", process_results(students))
```

# Problem 4: Cab Fare Estimator with Retry Option

## Problem Statement:

Create a cab fare calculator:

Base fare = ₹50

₹12 per km

Peak hour → 25% extra

Allow retry using while loop

## Real-Time Use:

Ride-hailing apps

Transport systems

## Hint:

Function for fare calculation

While loop for retry


```python
def calculate_fare(km, is_peak):
    base = 50
    fare = base + (12 * km)
    if is_peak:
        fare *= 1.25
    return fare

while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    
    fare = calculate_fare(distance, peak == "yes")
    print(f"Estimated Fare: ₹{fare:.2f}")``
    
    again = input("Calculate again? (yes/no): ").lower()
    if again != "yes":
        break
```

# Problem 5: Employee Attendance Eligibility Checker

## Problem Statement:
Create a function that:
Accepts attendance list ("P", "A").

Calculates attendance percentage

Returns Eligible if ≥ 75% else Not Eligible

## Real-Time Use:

HR payroll systems

Internship tracking

## Hint:

Use loop to count present days

Use return statement


```python
def check_attendance(att_list):
    present = 0
    total = len(att_list)

    for day in att_list:
        if day == "P":
            present += 1
    
    percentage = (present / total) * 100

    if percentage >= 75:
        return "Eligible"
    return "Not Eligible"

attendance = ["P", "A", "P", "P", "P", "A", "P"]
print("Attendance status:", check_attendance(attendance))
```

# Problem 6: Password Strength Checker

## Problem Statement:

Create a function to check password strength.

## Rules:

Minimum length = 8

Must contain at least one digit

Must contain at least one special character (@#$)

## Real-Time Use:

Login systems

Security validation

## Hint:

Use loop over string

Use conditions inside function

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Too short"
    
    digits = False
    special = False

    for ch in password:
        if ch.isdigit():
            digits = True
        if ch in "@#$":
            special = True

    if digits and special:
        return "Strong Password"
    return "Weak Password"

pw = input("Enter password: ")
print(check_password_strength(pw))
```
