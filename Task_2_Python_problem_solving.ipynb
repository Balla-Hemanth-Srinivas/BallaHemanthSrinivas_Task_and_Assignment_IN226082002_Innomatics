1. Unique words in a sentence

```python
sentence=input("Enter the sentence: ") # Taking user input
words=sentence.split() # Split the string ( can also use .lower().split() to avoid case sensitive issues )
unique=set(words) # Converting the List to Set to remove duplicates
print("Total Unique words :",len(unique))
print("Unique words in the sentence: ",unique)
```

2. Highest salary from Employee data

```python
emp={
    "Arun":55000,
    "Bhuvan":65000,
    "Chethan":75000
    } # Employee dictionary

top_emp=max(emp,key=emp.get) # Finding the employee with max salary
high_sal=emp[top_emp]
print(f"Highest salary: {top_emp} - {high_sal}")
```

3. Find Maximum and Minimum Values

```python
num=[45,22,89,10,66] # Defining the list
maximum=num[0] # Initialize mmax and min
minimum=num[0]
for n in num:
  if n > maximum:
    maximum=n
  if n < minimum:
    minimum=n
print(f"List: {num}")
print(f"Maximum value: {maximum}")
print(f"Minimum value: {minimum}")
```

4. Count Products above a price threshold

```python
prices=[450,1200,899,1500,300] # Defining the list
count=0
for p in prices:
  if p > 1000:
    count+=1
print(f"Products above 1000 : {count}")
```

5. Calculate Attendance percentage

```python
att=["P", "P", "A", "P", "P"] # List of attendance
p_count=att.count("P") # Count number of present
total_days=len(att) # Total number of days
att_percent=(p_count/total_days)*100 # calculate number of days
print(f"Attendance percentage : {att_percent}%")
```

6. Remove duplicate phone numbers

```python
ph_num=[9876543210, 9123456789, 9876543210] # List of phone numbers
uniq_ph=set(ph_num) # Converting to set to remove duplicates
print(f"Unique phone numbers: {uniq_ph}")
```

7. Count character frequency

```python
word="pyhtonpython" # Input string
freq={} # Initialize empty dictionary
for char in word:
  if char in freq:
    freq[char]+=1
  else:
    freq[char]=1
print(f"frequency: {freq}")
```

9. Check if a key exists in a dictionary

```python
emp={
  "Akshay": 65000,
  "Bhavesh": 75000,
  "Chethan": 85000
  } # Define dictionary

search_key="Akshay"
if search_key in emp:
  print("Employee exists")
else:
  print("Employee doesn't exist")
```

10. Calculate Average marks

```python
marks=[85,90,45,70] # Define List of marks
total_sum=sum(marks) # Calculate sum of all subjects
count=len(marks) # Find Total number of subjects
average=total_sum/count # Calculate average
print(f"Average marks: {average}")
```
